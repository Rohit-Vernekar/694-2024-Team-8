```python
import json
import pandas as pd 
```

```python
# report counts for corona-out-2

lines=0
tweets = {}
tweets_list = []
retweets_list = []
quoted_tweets_list = []
geo = 0
urls = 0
hashtags = 0
users = {}

with open("corona-out-2", "r") as f1:
    for line in f1:
        try:
            data = json.loads(line)
            lines = lines + 1

            if 'retweeted_status' in data.keys():
                retweets_list.append(data['id_str'])
                if (data['retweeted_status']['id_str'] not in tweets):
                    tweets[data['id_str']] = data
            
            if 'quoted_status' in data.keys():
                quoted_tweets_list.append(data['id_str'])
                if (data['quoted_status']['id_str'] not in tweets):
                    tweets[data['id_str']] = data
            
            tweets_list.append(data['id_str'])
            if (data['id_str'] not in tweets):
                tweets[data['id_str']] = data
            
        except:
            # if there is an error loading the json of the tweet, skip
            continue

df_tweets = pd.DataFrame({'tweet':tweets_list, 'tweeted':1}).drop_duplicates()
df_tweets['retweeted'] = df_tweets.tweet.isin(retweets_list)*1
df_tweets['quoted'] = df_tweets.tweet.isin(quoted_tweets_list)*1

for i in tweets:
    urls += min(len(tweets[i]['entities']['urls']),1)
    hashtags += min(len(tweets[i]['entities']['hashtags']),1)
    if tweets[i]['coordinates'] is not None:
        geo += 1
    if (tweets[i]['user']['id_str'] not in users):
        users[tweets[i]['user']['id_str']] = tweets[i]['user']
    
print('number of unique tweets/retweets/quoted:', df_tweets.shape[0], len(tweets))
print('number of unique tweets:',df_tweets[(df_tweets.retweeted == 0) & (df_tweets.quoted == 0)].shape[0])
print('number of unique retweets:', df_tweets.retweeted.sum())
print('number of unique quoted tweets:', df_tweets.quoted.sum() - df_tweets[(df_tweets.retweeted == 1) & (df_tweets.quoted == 1)].quoted.sum())
print('num of unique users=',len(users))
print('num of tweets with geo=',geo)
print('num of tweets with urls=',urls)
print('num of tweets with hashtags=',hashtags)
```
